The SpaceX Raptor is a cryogenic staged combustion rocket engine intended to power the high-performance lower and upper stages for the Interplanetary Transport System. It has more than three times the thrust of SpaceXs Merlin 1D engines propelling the Falcon 9 and Falcon Heavy rockets and steps away from a Kerosene-based propellant.Raptor consumes a combination of Liquid Methane and Liquid Oxygen in a Full-Flow Staged Combustion Cycle (see below).

The highly-reusable engine makes use of concepts first demonstrated on the Falcon 9 and Falcon Heavy rockets including deep cryogenics, cooled below their boiling point to increase their density and thus load the limited tank volume with a greater mass of propellant.

# Thrust and fuel consumption

For calculating thrust and fuel consumption, we can define functions for these calculations. The thrust calculation involves considering the throttle range, while the fuel consumption can be derived from the flow rates of liquid oxygen (LOX) and liquid methane (LCH4).



```python
class PropulsionSystem:
    def __init__(self, Designation, Type, Propellant_Feed, Oxidizer, Fuel,
                 Thrust_Sea_Level, Thrust_Vacuum, Specific_Impulse_SL, Specific_Impulse_Vac,
                 Chamber_Pressure, Throttle_Range, Ignition, Re_Start_Capability,
                 Area_Ratio, Mixture_Ratio, Flow_Rate_Calc, LOX_Flow_Rate, LCH4_Flow_Rate):
        self.Designation = Designation
        self.Type = Type
        self.Propellant_Feed = Propellant_Feed
        self.Oxidizer = Oxidizer
        self.Fuel = Fuel
        self.Thrust_Sea_Level = Thrust_Sea_Level
        self.Thrust_Vacuum = Thrust_Vacuum
        self.Specific_Impulse_SL = Specific_Impulse_SL
        self.Specific_Impulse_Vac = Specific_Impulse_Vac
        self.Chamber_Pressure = Chamber_Pressure
        self.Throttle_Range = Throttle_Range
        self.Ignition = Ignition
        self.Re_Start_Capability = Re_Start_Capability
        self.Area_Ratio = Area_Ratio
        self.Mixture_Ratio = Mixture_Ratio
        self.Flow_Rate_Calc = Flow_Rate_Calc
        self.LOX_Flow_Rate = LOX_Flow_Rate
        self.LCH4_Flow_Rate = LCH4_Flow_Rate

    def calculate_thrust(self, altitude=0):
        # Calculate thrust based on altitude and throttle range
        if altitude == 0:  # Sea level
            thrust = self.Thrust_Sea_Level * (self.Throttle_Range / 100)
        else:
            thrust = self.Thrust_Vacuum * (self.Throttle_Range / 100)
        return thrust

    def calculate_fuel_consumption(self):
        # Calculate fuel consumption based on flow rates
        fuel_consumption = self.LOX_Flow_Rate + self.LCH4_Flow_Rate
        return fuel_consumption

# Sample data
data = {
    'Designation': 'Raptor Sea-Level',
    'Type': 'Full-Flow Staged Combustion',
    'Propellant_Feed': 'Multi-Stage Turbopump',
    'Oxidizer': 'Sub-Cooled Liquid Oxygen',
    'Fuel': 'Sub-Cooled Liquid Methane',
    'Thrust_Sea_Level': 3050,
    'Thrust_Vacuum': 3297,
    'Specific_Impulse_SL': 334,
    'Specific_Impulse_Vac': 361,
    'Chamber_Pressure': 300,
    'Throttle_Range': 100,  # Assuming maximum throttle for simplicity
    'Ignition': 'Spark Igniter',
    'Re_Start_Capability': 'Yes',
    'Area_Ratio': 40,
    'Mixture_Ratio': 3.8,
    'Flow_Rate_Calc': 931.2,
    'LOX_Flow_Rate': 737.2,
    'LCH4_Flow_Rate': 194
}

# Create a PropulsionSystem object
propulsion_system = PropulsionSystem(**data)

# Calculate and display thrust
thrust_at_sea_level = propulsion_system.calculate_thrust(altitude=0)
thrust_at_altitude = propulsion_system.calculate_thrust(altitude=100)  # Example altitude: 100 km

print(f"Thrust at Sea Level: {thrust_at_sea_level} kN")
print(f"Thrust at Altitude: {thrust_at_altitude} kN")

# Calculate and display fuel consumption
fuel_consumption = propulsion_system.calculate_fuel_consumption()
print(f"Fuel Consumption: {fuel_consumption} kg/s")
```

# Raptor Fuel burn rate and Initial speed

To calculate the fuel burn rate and initial speed, we can use the equations involving thrust, specific impulse, and mass flow rate. The fuel burn rate is the mass flow rate of fuel consumed per unit time, and the initial speed is the velocity achieved by the rocket at the beginning of its trajectory.

```python
class PropulsionSystem:
    def __init__(self, Designation, Type, Propellant_Feed, Oxidizer, Fuel,
                 Thrust_Sea_Level, Thrust_Vacuum, Specific_Impulse_SL, Specific_Impulse_Vac,
                 Chamber_Pressure, Throttle_Range, Ignition, Re_Start_Capability,
                 Area_Ratio, Mixture_Ratio, Flow_Rate_Calc, LOX_Flow_Rate, LCH4_Flow_Rate):
        self.Designation = Designation
        self.Type = Type
        self.Propellant_Feed = Propellant_Feed
        self.Oxidizer = Oxidizer
        self.Fuel = Fuel
        self.Thrust_Sea_Level = Thrust_Sea_Level
        self.Thrust_Vacuum = Thrust_Vacuum
        self.Specific_Impulse_SL = Specific_Impulse_SL
        self.Specific_Impulse_Vac = Specific_Impulse_Vac
        self.Chamber_Pressure = Chamber_Pressure
        self.Throttle_Range = Throttle_Range
        self.Ignition = Ignition
        self.Re_Start_Capability = Re_Start_Capability
        self.Area_Ratio = Area_Ratio
        self.Mixture_Ratio = Mixture_Ratio
        self.Flow_Rate_Calc = Flow_Rate_Calc
        self.LOX_Flow_Rate = LOX_Flow_Rate
        self.LCH4_Flow_Rate = LCH4_Flow_Rate

    def calculate_fuel_burn_rate(self):
        # Calculate fuel burn rate based on the flow rate of liquid methane (LCH4)
        fuel_burn_rate = self.LCH4_Flow_Rate
        return fuel_burn_rate

    def calculate_initial_speed(self):
        # Calculate initial speed based on thrust, specific impulse, and mass flow rate
        thrust_at_sea_level = self.Thrust_Sea_Level * (self.Throttle_Range / 100)
        specific_impulse_at_sea_level = self.Specific_Impulse_SL

        initial_speed = (thrust_at_sea_level / (self.Flow_Rate_Calc * specific_impulse_at_sea_level))
        return initial_speed

# Sample data
data = {
    'Designation': 'Raptor Sea-Level',
    'Type': 'Full-Flow Staged Combustion',
    'Propellant_Feed': 'Multi-Stage Turbopump',
    'Oxidizer': 'Sub-Cooled Liquid Oxygen',
    'Fuel': 'Sub-Cooled Liquid Methane',
    'Thrust_Sea_Level': 3050,
    'Thrust_Vacuum': 3297,
    'Specific_Impulse_SL': 334,
    'Specific_Impulse_Vac': 361,
    'Chamber_Pressure': 300,
    'Throttle_Range': 100,  # Assuming maximum throttle for simplicity
    'Ignition': 'Spark Igniter',
    'Re_Start_Capability': 'Yes',
    'Area_Ratio': 40,
    'Mixture_Ratio': 3.8,
    'Flow_Rate_Calc': 931.2,
    'LOX_Flow_Rate': 737.2,
    'LCH4_Flow_Rate': 194
}

# Create a PropulsionSystem object
propulsion_system = PropulsionSystem(**data)

# Calculate and display fuel burn rate
fuel_burn_rate = propulsion_system.calculate_fuel_burn_rate()
print(f"Fuel Burn Rate: {fuel_burn_rate} kg/s")

# Calculate and display initial speed
initial_speed = propulsion_system.calculate_initial_speed()
print(f"Initial Speed: {initial_speed} m/s")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sample data
data = {
    'Specific Impulse (SL)': [334],
    'Chamber Pressure': [300],
    'Area Ratio': [40],
    'Mixture Ratio': [3.8],
    'Flow Rate (Calc.)': [931.2],
    'Throttle Range (%)': [100],
}

# Target variable
sea_level_thrust = 3050

# Creating a DataFrame
df = pd.DataFrame(data)
df['Sea Level Thrust'] = sea_level_thrust

# Splitting the data into features and target
X = df.drop('Sea Level Thrust', axis=1).values
y = df['Sea Level Thrust'].values

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Building a simple neural network
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_scaled.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model with the entire dataset
model.fit(X_scaled, y, epochs=100, batch_size=1, verbose=1)

# Making predictions on the same data
y_pred = model.predict(X_scaled)

# Evaluating the model
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
```

# Failure Detection and Abort Mission

failure occurs if the specific impulse in vacuum is below a certain threshold. In this project notebook, the check_failure method checks if the specific impulse in vacuum is below a threshold (set at 350 seconds for demonstration purposes).

# Raptor enging Failure rate

Creating a failure rate code typically involves defining specific failure conditions and then simulating multiple launches to determine the rate of failures. Since the specific failure conditions are based on the specific impulse in vacuum falling below a threshold.This python notebook uses a random range for the specific impulse in vacuum for each simulated launch.

```python
import random

class Rocket:
    def __init__(self, designation, specific_impulse_vac):
        self.designation = designation
        self.specific_impulse_vac = specific_impulse_vac

    def check_failure(self):
        # Placeholder for failure condition
        # This is a simplistic example where failure occurs if specific impulse in vacuum is below 350 seconds
        failure_threshold = 350
        if self.specific_impulse_vac < failure_threshold:
            return True
        else:
            return False

def simulate_launches(num_launches):
    # Simulate multiple launches and calculate failure rate
    failures = 0

    for _ in range(num_launches):
        # Generate random specific impulse in vacuum for each launch
        specific_impulse_vac = random.uniform(300, 400)  # Adjust the range based on your requirements

        # Create a Rocket object
        rocket = Rocket('Raptor Sea-Level', specific_impulse_vac)

        # Check for failure
        if rocket.check_failure():
            failures += 1

    failure_rate = failures / num_launches
    return failure_rate

# Simulate 100 launches (you can adjust the number based on your needs)
num_simulations = 100
failure_rate = simulate_launches(num_simulations)

print(f"Failure Rate: {failure_rate * 100:.2f}%")
```
